# Matplotlib - 다양한 플롯

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

1. 산점도(Scatter Plot)

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ans.groupby('dataset').agg(['mean', 'std'])
```

```python
X, Y = np.random.randn(100), np.random.randn(100)
plt.title('산점도')
plt.scatter(X, Y)
plt.show()
```

```python
colors = np.random.rand(100)
area = np.pi * (15 * np.random.rand(100))**2
plt.scatter(X, Y, s=area, c=colors, alpha=0.5)
plt.title('Bubble chart')
plt.show()
```

### 2. Bar Chart

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar chart')
plt.show()
```

```python
s.plot(kind='bar')
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s = titanic['class'].value_counts()
plt.bar(s.index, s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.barh(s.index, s.values)
plt.title('Bar chart')
plt.show()
```

### 3. Histogram

```python
s = np.random.randn(200)
plt.hist(s)
plt.title('Histogram')
plt.show()
```

```python
plt.hist(s, bins=20)
plt.title('bins=20인 경우')
plt.show()
```

```python
s = np.random.randn(20000)
plt.hist(s, bins=20)
plt.title('bins=20인 경우')
plt.show()
```
